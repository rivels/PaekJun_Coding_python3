```python
def addnmul(*args):
    su=0
    mul=1
    for i in args:
        su=su+i
    for j in args:
        mul=mul*j
    return su, mul

#함수에선 return을 만나면 바로 빠져나간다!
```

```python
def d():
    ten=one=hund=thou=0
    K=[] #생성되는 수들이 들어가는 리스트
    J=0  #생성되는 숫자가 담길 변수
    M=[]
    for i in range(0,10000):
        thou=i//1000            #4     #5    #
        hund=(i//100)%10        #3     #8
        ten=(i//10)%10          #6     #3
        one=i%10                       #7
        J=i=i+thou+hund+ten+one
        K.append(J)
    K=set(K)
    K=list(K)
    for j in range(1,10001):
        if j not in K:
            M.append(j)
    return M


'''
휴... 문제는 애초에 맞췄던 것 같은데 함수를 정의만 하고 함수를 실행하지 않았다
그래서 계속 오답이 나왔다....
흠... 다시는 까먹지 않을 것 같다^^

그리고 이 문제는 다소 무식한 방법으로 풀었는데
다른 더 효율적인 방법과 해답을 찾아보고 그에 맞게 다시 한번 코드를 작성해보자

오늘은 def를 이용한 함수에 대해서 공부했는데 
나름 재미있었다. 
'''
```

```python
def e():
    self_num = []
    M=[]
    for j in range(1, 10001):
        if j in self_num:
            new_num = j + sum([int(i) for i in list(str(j))])
            self_num.append(new_num)
        else:
            print(j)
            M.append(j)
            new_num = j + sum([int(i) for i in list(str(j))])
            self_num.append(new_num)
    return M
```
